```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(data=data,x="price")
```

From above plot since the histogram is not uniformly distributed i knew that there are some outliers which needs to be handled.First lets plot it using boxplot

```python
sns.boxplot(data=data,x='price')
```

Lets remove the overliers by taking into consideration only those values which are less than 1500

```python
df=data[data['price']<1500]
```

```python
sns.boxplot(data=df,x='price')
```

```python
df.columns
```

```python
df=df[df['number_of_reviews'] <350]
df=df[df["reviews_per_month"] <20]
```

```python
df.head()
```

Plotting Different Histograms Using for Loops

```python
columns=['price','availability_365','number_of_reviews','reviews_per_month']
plt.figure(figsize=(20,5* len(columns)))
for i,col in enumerate(columns,1):
    plt.subplot(len(columns),2,i)
    sns.histplot(data=df,x=col)
    plt.title(f'{col.capitalize()} Distribution')
    plt.ylabel("Frequency")
    plt.xlabel(f"{col}")
```

```python
df_grouped=df.groupby(by='neighbourhood_group')['price'].mean()
df_grouped.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(df_grouped,df_grouped.index,df_grouped.values)
plt.xlabel("Price")
plt.ylabel("Neighbourhood")
plt.title("Price per Neighbourhood")
```

Bronx is the most affordable borough.

```python
#feature engineering to find the price per bed
df['price_per_bed'] = df['price']/df['beds']
df['price_per_bed'].head()
```

```python
perbed_grouped=df.groupby(by="neighbourhood_group")["price_per_bed"].mean()
perbed_grouped.head()
```

```python
perbed_grouped.plot(kind='bar')
plt.xlabel("Neighbourhood")
plt.ylabel("Price per Bed")
plt.title("Price per Bed per Neighbourhood")
```

Manhattan is the most expensive per bed.

Brooklyn and Queens have moderate prices per bed.



```python
sns.barplot(data=df,x='neighbourhood_group',y='price',hue='room_type')
plt.title("Price per Neighbourhood and Room Type")
```

Entire homes are significantly more expensive than private or shared rooms.

Hotel rooms have high price variations.

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data=df,x="number_of_reviews",y="price",hue="neighbourhood_group")
plt.title("number_of_reviews vs price")
```

Listings with higher prices tend to have fewer reviews.

More affordable listings attract more reviews.

```python
sns.pairplot(data=df,vars=["price","availability_365","number_of_reviews","minimum_nights"],hue="room_type")
```

Helps visualize correlations and data distributions.

Shows trends between numerical variables.

```python
corr=df[["latitude","longitude","price","minimum_nights","number_of_reviews","availability_365","beds"]].corr()
corr
```

```python
sns.heatmap(data=corr,annot=True)
plt.title("Correlation Heatmap")
```

Price has a moderate correlation with the number of beds (0.41).

No strong correlation with other features.

```python
df_count=df['room_type'].value_counts()
df_count_dict=df_count.to_dict()
df_count_dict
```

```python
%pip install wordcloud
```

```python
df_count_dict = {str(k): v for k, v in df_count.items()}
```

```python
from wordcloud import WordCloud
```

```python
wordcloud=WordCloud(width= 500,height= 500,background_color='white').generate_from_frequencies(df_count_dict)
```

```python
plt.imshow(wordcloud)
plt.axis("off")
plt.title("Room Type Distribution")
```

Larger words indicate more frequent room types.

Private rooms and entire homes/apartments are the most common.

Shared rooms and hotel rooms are less frequent.
